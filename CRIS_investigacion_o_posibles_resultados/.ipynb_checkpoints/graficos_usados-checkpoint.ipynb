tal como hemos visto en el tema a tratar acerca de las carreras STEM (Ciencia, Tecnología, Ingeniería y Matemáticas), se presentará la visualización de datos de acerca matriculaciones anualmente en carreras STEM ya sea de que ingreso de primer año o como los que aún están en la carrera, y de la visualización de la brecha entre las matriculaciones en ingreso de primer año o como los que aún están en la carrera
Los datos presentados fueron sacados de directamente de informes anuales gubernamentales, se abarcará solo hasta 2023 debido que no hay datos de 2024.
(inflo: los datos totales refieren a los estudiantes que aun cursan las carreras de carácter STEM, y los de primer año matriculados refiere a los que recién ingresan a las carreras STEM)

A través de los gráficos se puede visualizar que si bien a lo largo del tiempo en los matriculados, y la brecha de genero entre ambos , se denota que:
- Que los matriculados de los hombres han aumentado mucho más que las mujeres en mantención de matrículas donde se pude ver una pequeña baja en el año 2020, donde según informe del Ministerio de Educación de Chile (2021), la disminución del -3,7% en la matrícula total en 2020 respecto a 2019 se debe al efecto conjunto del retraso del inicio del año académico por el estallido social y la irrupción de los primeros casos de COVID-19 en marzo y abril de 2020,
esto también se puede visualizar en los datos de los de primer año, donde se ve igualmente una baja en el año 2020, posiblemente por la misma razón, 
también denotar que las matriculas de las mujeres tanto los totales como los de primer año su aumento no se ha visto tan visible como al de su contraparte de los hombres, siendo que parece que pasar de los impulsos de universidades y el gobierno con sus programas de ingreso a carreras STEM sigue viéndose una gran brecha
Acerca del segundo grafico acerca de la brecha de genero se presenta esa brecha del cual se podía visualizar en el grafico de matriculados (ósea el primer gráfico)
tanto los de primer año como en los matriculados totales, siendo la baja más destacada en el año 2013, también denota que entre 2014 a 2021 ha habido una baja en la brecha al menos en términos de las matrículas totales, esto se  deduce a las posibles programas de ya sea las universidades y del gobierno (como el +MC que abre cupos adicionales en carreras STEM), o de empresas sin fines de lucro como la Fundación Luksic ofrece becas para mujeres en carreras técnicas STEM, cubriendo hasta cinco semestres de estudios. Además, proporciona programas de mentoría y capacitación, organiza eventos y conferencias, y forma alianzas estratégicas para fomentar la inclusión y el éxito de mujeres en áreas de ciencia, tecnología, ingeniería y matemáticas.
El crecimiento mayor magnitud también se ve en los datos de hombres sea primer año o total, esto se ve reforzado según el grafico de regresión lineal propuesto, también se verifica que el crecimiento de matriculados en mujeres aun no es el mejor para ir reduciendo la brecha en base a los hombres, demostrando aun así un largo camino de trabajar
