```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

# Beneficiarios ANID

Cantidad de proyectos beneficiarios de ANID, distribución porcentual y brecha de género, según sexo del responsable del proyecto, años 2015 a 2022.

```python
path_anid = os.path.join('data', 'excel', 'anid.xlsx')

df_anid = pd.read_excel(path_anid, skiprows=2)
df_anid = df_anid.iloc[:8]
df_anid = df_anid.drop(columns='Proyectos beneficiarios liderados por hombres o mujeres\n (Total)')
df_anid
```

```python
columnas = ['Año', 'Beneficiarios por Hombres', 'Beneficiarios por Mujeres', 'Porcentaje Beneficiarios por Hombres', 'Porcentaje Beneficiarios por Mujeres', 'Brecha de Genero']
df_anid.columns = columnas
df_anid.info()
```

```python
df_anid = df_anid.round(3)
df_anid['Año'] = df_anid['Año'].astype(int)
df_anid
```

```python
plt.figure(figsize=(12, 8))
plt.plot(df_anid['Año'], df_anid['Beneficiarios por Hombres'], marker='o', label='Hombres', color='blue')
plt.plot(df_anid['Año'], df_anid['Beneficiarios por Mujeres'], marker='x', label='Mujeres', color='pink')

plt.title("Beneficiarios de ANID por género a lo largo de los años")
plt.xlabel("Año")
plt.ylabel("Número de Beneficiarios")
plt.xticks(df_anid['Año'])
plt.legend(loc='best')
plt.grid(alpha=0.3)
plt.tight_layout()

plt.show()
```

# Embarazo No Planificado

Porcentaje de mujeres y hombres jóvenes iniciados sexualmente que vivieron un embarazo no planificado, según sexo y año.

```python
path_noplanificado = os.path.join('data', 'excel', 'embarazo_noplanificado.xlsx')

df_noplanificado = pd.read_excel(path_noplanificado, skiprows=2)
df_noplanificado = df_noplanificado.iloc[:7]
df_noplanificado = df_noplanificado.drop(columns='Unnamed: 1')
df_noplanificado
```

```python
columnas = ['Año', 'Porcentaje de Mujeres %', 'Porcentaje de Hombres %', 'Brecha']
df_noplanificado.columns = columnas
df_noplanificado = df_noplanificado.dropna()
df_noplanificado['Año'] = df_noplanificado['Año'].astype(int)
df_noplanificado['Porcentaje de Mujeres %'] = df_noplanificado['Porcentaje de Mujeres %'].astype(float)
df_noplanificado['Porcentaje de Hombres %'] = df_noplanificado['Porcentaje de Hombres %'].astype(float)
df_noplanificado['Brecha'] = df_noplanificado['Brecha'].astype(float)
df_noplanificado
```

```python
plt.figure(figsize=(12, 8))
plt.plot(df_noplanificado['Año'], df_noplanificado['Porcentaje de Hombres %'], marker='o', label='Hombres', color='blue')
plt.plot(df_noplanificado['Año'], df_noplanificado['Porcentaje de Mujeres %'], marker='x', label='Mujeres', color='pink')

plt.title("Porcentaje de mujeres y hombres jóvenes que vivieron un embarazo no planificado entre los 15 y 29 años")
plt.xlabel("Año")
plt.ylabel("Número de jóvenes")
plt.xticks(df_noplanificado['Año'])
plt.legend(loc='best')
plt.grid(alpha=0.3)
plt.tight_layout()

plt.show()
```
